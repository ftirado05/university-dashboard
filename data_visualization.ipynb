```python
# Import basic libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Upload the CSV file
df = pd.read_csv('university_student_data.csv')

# View the first few rows and general information
df.head()
df.info()
```

```python
# Retention rate trend over time

plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Year', y='RetentionRate', marker='o')
plt.title('Retention Rate Over Time')
plt.ylabel('Retention Rate (%)')
plt.xlabel('Year')
plt.show()
```

```python
# Average satisfaction per year

avg_satisfaction = df.groupby('Year')['Student Satisfaction (%)'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=avg_satisfaction, x='Year', y='Student Satisfaction (%)')
plt.title('Average Satisfaction Score by Year')
plt.ylabel('Satisfaction Score')
plt.xlabel('Year')
plt.show()
```

```python
# Comparison between terms (Spring vs Fall)

term_comparison = df.groupby('Term')['Enrolled'].sum().reset_index()

plt.figure(figsize=(6,6))
plt.pie(term_comparison['Enrolled'], labels=term_comparison['Term'], autopct='%1.1f%%')
plt.title('Enrollment by Term')
plt.show()
```
